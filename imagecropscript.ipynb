```python
# CHATGPT 2
import os
import pandas as pd
import cv2

# your directories
INPUT_DIR  = r"C:\Users\dtaylor\Documents\Jostle Pictures Project\ALLIDS"
OUTPUT_DIR = r"C:\Users\dtaylor\Documents\Jostle Pictures Project\jostleupload3"
CSV_PATH   = r"C:\Users\dtaylor\Documents\Jostle Pictures Project\jostleAllUsers_WE_FN_LN.csv"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# load your DataFrame
df = pd.read_csv(CSV_PATH, encoding='utf-8', low_memory=False)

# 1) Strip whitespace off all column names
df.columns = df.columns.str.strip()

# 2) Rename columns if they’re not already 'FirstName'/'LastName'
#    (adjust the left-hand keys to whatever your CSV actually uses)
df.rename(columns={
    'First Name': 'FirstName',
    'Last Name':  'LastName',
    'first_name': 'FirstName',
    'last_name':  'LastName',
}, inplace=True)

print("Columns in DataFrame now:", df.columns.tolist())

def crop_and_save(img_path: str, email: str, file_ext: str):
    img = cv2.imread(img_path)
    if img is None:
        print(f"[ERROR] failed to load image: {img_path}")
        return
    crop_img = img[563:1150, 500:960]
    out_path = os.path.join(OUTPUT_DIR, f"{email}{file_ext}")
    if cv2.imwrite(out_path, crop_img):
        print(f"[OK] cropped & saved → {out_path}")
    else:
        print(f"[ERROR] could not save to {out_path}")

for fname in os.listdir(INPUT_DIR):
    low = fname.lower()
    if not (low.endswith(".jpg") or low.endswith(".jpeg") or low.endswith(".png")):
        continue

    base, ext = os.path.splitext(fname)
    print(f"\nProcessing file: {fname!r}")

    # Extract first and last name:
    if "," in base:
        last_name, rest = base.split(",", 1)
        last_name = last_name.strip()
        first_name = rest.strip().split()[0]
    elif " " in base:
        parts     = base.split()
        first_name = parts[0].strip()
        last_name  = parts[-1].strip()
    else:
        print(f"[SKIP] can't parse name from {base!r}")
        continue

    print(f"  Parsed FirstName={first_name!r}, LastName={last_name!r}")

    # ---- Now this will succeed because we've ensured the cols exist ----
    mask  = (df['FirstName'] == first_name) & (df['LastName'] == last_name)
    count = mask.sum()

    if count == 1:
        row   = df.loc[mask].iloc[0]
        email = row['WorkEmail']
        print(f"[MATCH] {first_name} {last_name} → {email}")
        crop_and_save(
            os.path.join(INPUT_DIR, fname),
            email,
            ext
        )
    elif count > 1:
        print(f"[AMBIGUOUS] {count} matches for {first_name} {last_name}")
    else:
        print(f"[NOT FOUND] no match for {first_name} {last_name}")
```
